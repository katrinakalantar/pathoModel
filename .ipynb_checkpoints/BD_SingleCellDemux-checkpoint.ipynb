```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from collections import Counter

%matplotlib inline
```

```python
counts = pd.read_csv("/Users/kkalantar/Downloads/bHR40_untx.bd_demux_tagReadCounts.csv")
#counts = pd.read_csv("/Users/kkalantar/Downloads/bHR40_untx.bd_demux_ratioLow.cell2Label.csv")
```

```python
X = counts.fillna(0)
X = np.log10(X[['SampleTag_03|stAbO','SampleTag_04|stAbO']]+1)  # log-transform counts

# run k-means to establish original groups
kmeans = KMeans(n_clusters=3, random_state=1).fit(X)
```

```python
colors = ['red','orange','yellow','green','blue','purple']

plt.figure(figsize=[6,6])
plt.scatter(X[['SampleTag_03|stAbO']],X[['SampleTag_04|stAbO']], color=[colors[i] for i in kmeans.labels_], alpha = .5)

X['col'] = kmeans.labels_  # add kmeans labels to the matrix 
```

```python
# you have to fill this value in based on the color of the bottom left cluster!
# red = 0
# orange = 1
# yellow = 2
cluster_to_remove = 1
X = X[X['col'] != cluster_to_remove]  # remove cells from the bottom left corner

num_original_cells = counts.shape[0]
num_removed_cells = counts.shape[0] - X.shape[0]
print("N = " + str(num_removed_cells) + " cells removed from the analysis (lower left corner)")
```

```python
# run kmeans on the remaining samples
kmeans = KMeans(n_clusters=3, random_state=None).fit(X[['SampleTag_03|stAbO','SampleTag_04|stAbO']])

plt.figure(figsize=[6,6])
plt.scatter(X[['SampleTag_03|stAbO']],X[['SampleTag_04|stAbO']], 
            color=[colors[i] for i in kmeans.labels_], alpha = .5)

# compute the percentage of cells in each final cluster
ser = pd.Series(Counter(kmeans.labels_))
ser/num_original_cells
```
